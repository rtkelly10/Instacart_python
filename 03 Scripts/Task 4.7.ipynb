```python
#create subset
df = df_ords_prods_merged[:1000000]
```

```python
#check shape
df.shape
```

```python
#Define price-label function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Count unique values
df['price_range'].value_counts(dropna = False)
```

```python
#Check max
df['prices'].max()
```

#### If-Statements with the loc() Function

```python
#create conditions
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Check value counts
df['price_range_loc'].value_counts(dropna = False)
```

```python
#if statement for question
hour_of_order = []

for value in df_ords_prods_merged["hour_of_order"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hour_of_order.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    hour_of_order.append("Fewest orders")
  else:
    hour_of_order.append("Average orders")
```
